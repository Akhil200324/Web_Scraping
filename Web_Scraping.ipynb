```python
#Extract listings of the entire page
#below raisesAttribute Error so i need go for iterators -->for
#soup.find_all('div',attrs={'class':'title-line'}).text
```

```python
d = soup.find_all('div',attrs={'class':'title-line'})
print(len(d))
```

```python
e = []
for i in d:
    #print(i.text)
    e.append(i.text)
#print(e)
```

```python
#Comprehension for above one
prop_name = [i.text for i in d]
#print(prop_name)
```

```python
#Price for all listings in frst page
f = soup.find_all('div',attrs={'data-type':'price-link'})
price = [i.text.strip() for i in f]
print(price)
```

```python
#Location name
k = soup.find_all('span',attrs={'itemprop':'addressLocality'})
location = [i.text for i in k]
print(location)
```

```python
#Area listings
g = soup.find_all('td',attrs={'class':'size'})
area = [i.text.strip() for i in g]
print(area)
print(len(area))
```

```python
#Price per sqft
h = soup.find_all('td',attrs={'class':'lbl rate'})
prc_sqft = [i.text for i in h]
print(prc_sqft)
print(len(prc_sqft))
```

```python
#Status of construction
i =soup.find_all('td',attrs={'class':'val'})
cons = [i.text for i in i]
print(cons)
print(len(cons))
```

```python
#Builder name
h = soup.find_all('a',attrs={'class':'seller-name'})
builder_name = [i.text for i in h]
print(builder_name)
print(len(builder_name))
```

```python
#Listingsof bathrooms -->differnet
o = soup.find_all('li',attrs={'class':'keypoint',
                     'title':'Bathrooms'})
q = [i.text for i in o]
print(len(q))
print(q)
```

```python
#As in my case 19 listings are missing -->will fill them
s = soup.find_all('ul',attrs={'class':'listing-details'})
nbath_info = []
for i in s:
    if i.find('li',attrs={'class':'keypoint',
                     'title':'Bathrooms'})is None:
        nbath_info.append('Not Mentioned')
    else:
        d=i.find('li',attrs={'class':'keypoint',
                     'title':'Bathrooms'})
        nbath_info.append(d.text)
print(nbath_info)
```

```python
#Final step is storing the data in table format
data = pd.DataFrame({'Name':prop_name,
                    'Area':area,
                    'Price':price,
                    'Builder_Name':builder_name,
                    'Location':location,
                    'Bathrooms':nbath_info},
                   index=list(range(1,21)))
data
#finally store in a csv file
data.to_csv('Hyd_Apmnts.csv')
```
